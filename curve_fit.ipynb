```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'dis_rssi_data/ac233f66c6a2.csv'
```

```python
df = pd.read_csv(filename,names=['dist','rssi'])
```

```python
df = df.drop_duplicates()
```

```python
plt.scatter(df['dist'],df['rssi'])
```

```python
df1 = df[df.dist!=0]
```

```python
def func(x,a,b):
    return a*np.log10(x)+b
```

```python
from scipy.optimize import curve_fit
```

```python
popt,pcov = curve_fit(func, df1['dist'], df1['rssi'], maxfev=5000)
```

```python
plt.plot(df1['dist'],func(df1['dist'],*popt),'r-')
plt.scatter(df['dist'],df['rssi'])
plt.title('ac233f66c914:  '+str(round(popt[0],5))+'log(x) '+str(round(popt[1],5)))
plt.xlabel("distance")
plt.ylabel("RSSI")
```
